```python
%%sql

SELECT 
    Year, 
    Total, 
    Total - LAG(Total, 1) OVER (ORDER BY Year) AS YoY_Difference
FROM 
    suicides_india
LIMIT 50;
```

This explains the year to year difference in the data set. 
However, what was seeen in the output is because there are repitition of years.

```python
%%sql

SELECT 
    Year, 
    Total, 
    LAG(Total, 1) OVER (ORDER BY Year) AS Prev_Year_Total, 
    LEAD(Total, 1) OVER (ORDER BY Year) AS Next_Year_Total
FROM 
    suicides_india
LIMIT 100;
```

This is a leading and lagging analysis showing the previous and the following (next) year.
